## Prevendo o preço da pizza

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Vamos supor que você registrou o tamanho e preço da pizza que você comeu nos ultimos meses com a sua familia
Diametro(cm)  |  Preço

              7      8
             10     11
             15     16
             30   38.5
             45     52 


```python
# Diametro (cm)
diametros = [[7], [10], [15], [30], [45]]

# Preço (R$)
precos = [[8], [11], [16], [38.5], [52]]
```

```python
# Vamos visualizar os dados construindo um plot
plt.figure()
plt.xlabel('Diametro (cm)')
plt.ylabel('Preço (R$)')
plt.title('Diametro x Preço')
plt.plot(diametros, precos, 'k.')
plt.axis([0, 60, 0, 60])
plt.grid(True)
plt.show()
```

```python
# importando o modulo de Regressão Linear do scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
# Preparando os dados de treino

# Vamos chamar de X os dados de diametro da Pizza
X = [[7], [10], [15], [30], [45]]

# Vamos chamar de Y os dados de preço da Pizza
Y = [[8], [11], [16], [38.5], [52]]
```

```python
# Criando o modelo
modelo = LinearRegression()
```

```python
# Treinando o modelo
modelo.fit(X, Y)
```

```python
# Prevendo o preço de uma pizza de 20 cm de diametro
print("Uma pizza de 20 cm de diametro deve cusar: R$%.2f" % modelo.predict(np.array([20][0]).reshape(1, 1)))
```

## Construindo o Scatter PLot

```python
# Coeficientes
print('Coeficiente: ', modelo.coef_)

# MSE (mean square error)
print('MSE: ', np.mean((modelo.predict(X) - Y) ** 2))

# Score de variação: 1 - Representa a predição perfeita (100% de acuracia)
print('Score de variação ', modelo.score(X, Y))
```

```python
# Scatter Plot representando a regressão linear
plt.scatter(X, Y, color = 'black')
plt.plot(X, modelo.predict(X), color = 'blue', linewidth = 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks(())
plt.yticks(())
plt.show()
```
