```python
import matplotlib.pyplot as plt
import matplotlib as mpl 
import random
import numpy as np
%matplotlib inline
```

```python
# o metodo plot define os eixos do grafico
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
x = [1, 4, 5]
y = [6, 7, 8]
```

```python
plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Grafico X Y')
plt.show()
```

```python
x2 = [1, 2, 3]
y2 = [12, 13, 14]
x3 = [1, 6, 10]
y3 = [3, 16, 30]
plt.plot(x2, y2, label = 'Primeira Linha')
plt.plot(x3, y3, label = 'Segunda Linha')
plt.legend()
plt.show()
```

```python
# Barras
x = [2, 3, 4, 5, 6]
y = [5, 7, 8, 9, 10]
plt.bar(x, y, label = 'Barras', color = 'red')
plt.legend()
plt.show()
```

```python
x2 = [1, 2, 3, 5, 6]
y2 = [2, 3, 4, 5, 6]
plt.bar(x, y, label = 'Barra 01', color = 'red')
plt.bar(x2, y2, label = 'Barra 02', color = 'blue')
plt.legend()
plt.show()
```

```python
idades = [22, 33, 43, 55, 67, 87, 12, 22, 32, 43, 26, 28, 40, 50, 59, 64, 52]
ids = [x for x in range(len(idades))]
plt.bar(ids, idades)
plt.show()
```

```python
# Histograma
bins = list(range(0, 131, 10))
plt.hist(idades, bins, histtype = 'bar', rwidth = 0.8)
plt.show()
```

```python
plt.hist(idades, bins, histtype = 'stepfilled', rwidth = 0.8)
plt.show()
```

```python
# Scatterplot
x = list(range(1, 9, 1))
y = [5, 2, 4, 5, 6, 8, 4, 8]
plt.scatter(x, y, label = 'Pontos', color = 'red', marker = 'o', s = 100)
plt.legend()
plt.show()
```

```python
# stack plots
dias = [1, 2, 3, 4, 5]
dormir = [7, 8, 6, 77, 7]
comer = [2, 3, 4, 5, 3]
trabalhar = [7, 8, 9, 7, 6]
passear = [8, 5, 6, 7, 9]
plt.stackplot(dias, dormir, comer, trabalhar, passear, colors = ['m', 'c', 'r', 'k', 'b'])
plt.show()
```

```python
# Pie Char
fatias = [7, 2, 3, 6]
atividades = ['Engenheiro', 'Cientista', 'Analista', 'Arquiteto']
cores = ['r', 'm', 'b', 'k']
plt.pie(fatias, labels = atividades, colors = cores, startangle = 90, shadow = True, explode = (0, 0.1, 0, 0))
plt.show()
```

```python
# Graficos a partir do Numpy
plt.scatter(np.arange(0, 50), np.random.randn(50))
plt.show()
```

```python
# plot a scatter
fig = plt.figure()

axi1 = fig.add_subplot(1, 2, 1)
axi1.plot(np.random.randn(50), color = 'red')

axi2 = fig.add_subplot(1, 2, 2)
axi2.scatter(np.arange(50), np.random.randn(50))

plt.show()
```

```python
# plots diversos
_, ax = plt.subplots(2, 3)

ax[0, 1].plot(np.random.randn(50), color = 'green', linestyle = '-')
ax[1, 0].hist(np.random.randn(50))
ax[1, 2].scatter(np.arange(50), np.random.randn(50), color = 'red')
plt.show()
```
