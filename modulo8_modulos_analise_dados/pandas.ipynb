```python
# Time Series - Gerando por Segundos
rng = pd.date_range('1/1/2018', periods = 50, freq = 'S')
ts = pd.Series(np.random.randint(0, 500, len(rng)), index = rng)
ts
```

```python
# Gerando por Mês
rng = pd.date_range('1/1/2018', periods = 50, freq = 'M')
ts = pd.Series(np.random.randint(0, 500, len(rng)), index = rng)
ts
```

```python
# Plotting
import matplotlib.pyplot as plt 
from matplotlib import style
%matplotlib inline
```

```python
#Time Series Plot
ts = pd.Series(np.random.randn(500), index = pd.date_range('1/1/2016', periods = 500))
ts = ts.cumsum()
ts.plot()
```

```python
# Dataframe Plot
df = pd.DataFrame(np.random.randn(500, 4), index = ts.index, columns = ['A', 'B', 'C', 'D'])
df = df.cumsum()
plt.figure(); df.plot(); plt.legend(loc = 'best')
```
