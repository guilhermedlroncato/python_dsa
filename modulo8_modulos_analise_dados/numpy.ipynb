```python
a = np.array([15, 23, 40, 50, 90])
```

```python
# Em estatica a media é o valor que aponta para onde mais se concentram os dados de uma distribuição
np.mean(a)
```

```python
# O desvio padrão mostra o quanto de variação ou dispersão existem em relação a media (ou valor esperado). Um baixo desvio padrão indica que os dados tendem a estar proximos da media e um alto desvio padrão, indica que os dados estão espalhados por uma gama de valores
np.std(a)
```

```python
# Variancia de uma variavel aleatoria é uma media de sua dispersao estatistica, indicando o quão longe em geral os seus valores se encontram do valor esperado
np.var(a)
```
